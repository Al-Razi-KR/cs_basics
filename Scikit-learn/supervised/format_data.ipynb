```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# Load dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['species'] = data.target
df.head()
```

```python
# Arrange Data into Features Matrix and Target  Vector
feature_names = [
    'sepal length (cm)',
    'sepal width (cm)',
    'petal length (cm)',
    'petal width (cm)',
]
```
